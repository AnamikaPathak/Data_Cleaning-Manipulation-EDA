###  MISSING VALUES TREATMENT

* As per the above analysis we can see that there are a lot of missing values present in the variables. Here we can do 3 things to treat these missing values: 
    * 1. We can drop those columns where more than 50% of data is missing. Reason behind that, while doing predictive modeling, less no. of data can lead to less accurate prediction. 
    * 2. Continuous Variable - We can fill the null values with mean or median, as very less no. of data (around 6-8%) is missing in few variables. 
    * 3. Categorical Variables - We can fill the null values with the most frequent value present in that particular variable.

```python
df.drop(columns = ['ListingNumber', 'ListingNumber','ListingCreationDate','CreditGrade', 'ClosedDate','ProsperRating (numeric)', 'GroupKey', 'TotalProsperPaymentsBilled', 
                   'TotalProsperLoans', 'OnTimeProsperPayments','ProsperPaymentsLessThanOneMonthLate',
                   'ProsperPaymentsOneMonthPlusLate','ProsperPrincipalBorrowed','ProsperPrincipalOutstanding',
                   'ScorexChangeAtTimeOfListing', 'LoanFirstDefaultedCycleNumber', 'FirstRecordedCreditLine', 
                   'IncomeVerifiable'], inplace = True)
```

```python
#converting data type to remove boolean variables 
#IsBorrowerHomeowner 
df ['IsBorrowerHomeowner'] = df ['IsBorrowerHomeowner'].map({True:'Yes', False: 'No'})

# CurrentlyInGroup
df ['CurrentlyInGroup'] = df ['CurrentlyInGroup'].replace(to_replace=False, value="No")
df ['CurrentlyInGroup'] = df ['CurrentlyInGroup'].replace(to_replace=True, value="Yes")
```

```python
df['CurrentlyInGroup'].value_counts()
```

```python
df.info()
```

```python
# Separating numical and categorical data
numerical_var = df.dtypes[df.dtypes != "object"].index
data_num = df[numerical_var]

categorical_var = df.dtypes[df.dtypes == "object"].index
data_cat = df[categorical_var]
```

```python
# TREATING MISSING VALUES OF NUMERICAL COLUMNS
data_num = data_num.fillna(data_num.median())
data_num
```

```python
# TREATING MISSING VALUES OF CATEGORICAL COLUMNS

#ProsperRating (Alpha) 
data_cat['ProsperRating (Alpha)'] = data_cat['ProsperRating (Alpha)'].fillna('C')
```

```python
#BorrowerState
data_cat ['BorrowerState'] = data_cat ['BorrowerState'].fillna ('CA')
```

```python
#Occupation 
data_cat ['Occupation'] = data_cat ['Occupation'].fillna ('Other')

data_cat['Occupation'] = data_cat['Occupation'].replace (['Student - Technical School', 'Student - College Freshman', 
                                                         'Student - Community College', 'Student - College Sophomore',
                                                         'Student - College Sophomore', 'Student - College Junior', 
                                                'Student - College Senior','Student - College Graduate Student'],'Students')

data_cat['Occupation'] = data_cat['Occupation'].replace (['Computer Programmer', 'Engineer - Mechanical', 'Engineer - Chemical',
                                                         'Engineer - Electrical'], 'Engineers')

data_cat['Occupation'] = data_cat['Occupation'].replace (['Pharmacist', 'Chemist'], 'Pharmacist')
data_cat['Occupation'] = data_cat['Occupation'].replace (['Doctor', 'Psychologist', 'Biologist', 'Nurse (LPN)', 'Nurse (RN)'], 
                                                         'Medical Professionals')

data_cat['Occupation'] = data_cat['Occupation'].replace (['Professor', 'Teacher'], 'Teaching Professionals')
```

```python
# EmploymentStatus
data_cat ['EmploymentStatus'] = data_cat ['EmploymentStatus'].fillna ('Employed')
```

From the above statistical analysis we can see that there are maximum outliers present in the 3rd quantile which needs to be removed. 

### Outlier Treatment 

```python
def outlier_capping(x):
    x = x.clip(upper = x.quantile(0.99))
    x = x.clip(lower = x.quantile(0.01))
    return x
```

```python
data_num=data_num.apply(outlier_capping)
```

```python
df = pd.concat([data_num , data_cat], axis = 1)
df
```
