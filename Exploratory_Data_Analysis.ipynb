* CREDIT SCORE 

```python
# Make my cluster
def Credit_score(x):
    outcome=""
    if x >= 300 :
        outcome = "Poor"
    if x>=550:
        outcome = "Fair"
    if x>=650:
        outcome = "Good"
    if x>=750:
        outcome = "Excellent"
    return outcome
```

```python
Data ['CreditScoreRangeUpper'] = Data ['CreditScoreRangeUpper'].apply(Credit_score)
```

```python
Data ['CreditScoreRangeUpper'].value_counts().plot.bar(figsize=(10,5), color = 'skyblue') 
```

```python
def barPerc(df,xVar,ax):
    numX=len([x for x in df[xVar].unique() if x==x])
    bars=ax.patches
    for ind in range(numX):
        hueBars=bars[ind:][::numX]
        total=sum([x.get_height() for x in hueBars])
        for bar in hueBars:
            ax.text(bar.get_x() + bar.get_width()/2.,bar.get_height(),f'{bar.get_height()/total:.0%}',ha="center",va="bottom")
        
```

```python
# LoanStatus vs IsBorrowerHomeOwner

plt.figure(figsize=[15,5])
ax=sns.countplot(x='LoanStatus',hue='IsBorrowerHomeowner',data=Data)
barPerc(Data,'LoanStatus',ax)
```
